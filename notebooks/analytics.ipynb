# Business Sales Analytics Project

This project analyzes retail sales data using PostgreSQL, SQL, Python, and Pandas to generate business insights and KPIs.

## Problem Statement

The goal of this project is to analyze sales transactions and identify:
- top-performing categories
- best-selling products
- regional performance
- customer behavior
- business KPIs

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
DATABASE_URL = "postgresql://postgres:123456@localhost:5432/sales_analytics"

engine = create_engine(DATABASE_URL)
```

```python
query = "SELECT * FROM sales"

df = pd.read_sql(query, engine)

df.head()
```

## Dataset Overview

The dataset contains:
- transaction information
- customer demographics
- product details
- regional sales data

```python
df.columns
```

## Key Performance Indicators (KPIs)

This section analyzes:
- total sales
- average sales
- category performance
- regional performance

```python
total_sales = df["sales_amount"].sum()

round(total_sales, 2)
```

## Category Sales Analysis

This analysis identifies the highest revenue-generating product categories.

```python
category_sales = (
    df.groupby("category")["sales_amount"]
    .sum()
    .sort_values(ascending=False)
)

category_sales
```

## Observations

- Electronics category generates the highest sales.
- Sports products show strong customer demand.
- Some categories contribute significantly less revenue.
